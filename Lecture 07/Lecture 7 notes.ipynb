(d) Create a loop that ...


- Goes over the list of years
- If the year $\ge$ 2010, print the message

$\qquad$ "This dataset contains information on or after 2010.

- If the year $<$ 2010, print the message

$\qquad$ "This dataset contains information before 2010"

```python
wdi_2000 = pd.read_csv('wdi_data/wdi_2000.csv')
wdi_2020 = pd.read_csv('wdi_data/wdi_2020.csv')
year1 = [wdi_2000['year'],wdi_2020['year']]

for years in year1:
    for x in years:
        if x>=2010:
            print("This dataset contains information on or after 2010")
        elif x<2010:
            print("This dataset contains information before 2010")
```

```python

wdi_data = pd.read_csv('wdi_data/wdi_2000.csv')
wdi_data.to_excel('wdi_data/wdi_2000.xlsx')

a=['wdi_data/wdi_2000.csv','wdi_data/wdi_2020.csv']
excel_datasets=['wdi_data/wdi_2000.xlsx','wdi_data/wdi_2020.xlsx']
index=0
for x in a:
    wdi = pd.read_csv(x)
    wdi.to_excel(excel_datasets[index])
    index+=1



```

(f) Create a loop that ...

- Reads multiple ".csv" files
- Creates a numeric vector by 
    - Adding/Subtracting/Mutiplying/Diving two or more variables 
- Plots a histogram of the transformed variable
    - Numbers the figures
    - Labels the axes
    - Labels the title

```python
a=['data/features.csv','data/features_stored.csv']

for x in a:
    wdi = pd.read_csv(x)
    num_vec=wdi['acceleration']/wdi['cylinders']*wdi['mpg']+wdi['displacement']
    plt.hist(x=num_vec)
    plt.title('A/W')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
    
```
